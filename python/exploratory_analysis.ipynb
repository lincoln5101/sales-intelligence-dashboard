# Exploratory Data Analysis — Sales Intelligence Dashboard

**Purpose:** Explore raw sales data before loading into SQL. Identify patterns, outliers, and data quality issues that inform the cleaning pipeline.

**Run after:** Placing the raw dataset in `data/raw/`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

PROJECT_ROOT = Path('..').resolve()
RAW_DIR = PROJECT_ROOT / 'data' / 'raw'

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# Load raw data (update filename after dataset selection)
csv_files = list(RAW_DIR.glob('*.csv'))
if not csv_files:
    raise FileNotFoundError('No CSV found in data/raw/. Add your dataset first.')

df = pd.read_csv(csv_files[0])
print(f'Loaded: {csv_files[0].name}')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Data types and missing values
df.info()
df.isnull().sum()
```

```python
# Summary statistics for numeric columns
df.describe()
```

## Revenue & Profit Analysis

```python
# Monthly revenue trend (update column names if needed)
if 'Order Date' in df.columns and 'Sales' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'])
    monthly = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
    monthly.plot(kind='bar', title='Monthly Revenue')
    plt.ylabel('Revenue ($)')
    plt.tight_layout()
    plt.show()
```

```python
# Profit margin by category
if 'Category' in df.columns and 'Sales' in df.columns and 'Profit' in df.columns:
    cat = df.groupby('Category').agg(revenue=('Sales', 'sum'), profit=('Profit', 'sum'))
    cat['margin_pct'] = (cat['profit'] / cat['revenue'] * 100).round(1)
    cat.sort_values('revenue', ascending=False)
```
